<h1> <strong> 1 - Signal pure

<h2> Considérons le signal pure qui suit

```python
n = 5000
t = np.linspace(0,1,n)

f_0 = 2.*np.sin(2.*np.pi*70*t) + 3.5*np.cos(2.*np.pi*10*t) - 4*np.sin(2.*np.pi*20*t)
```

<h2> Représentation graphique du signal

```python
plt.figure(figsize=(10,7))
plt.plot(t , f_0 , color ="c", lw = 1.6, label="signal pure" )

plt.title("Signal pure", size = 16)
plt.legend()
plt.show()
```
